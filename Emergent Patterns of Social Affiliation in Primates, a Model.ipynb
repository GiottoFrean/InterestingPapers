# Emergent Patterns of Social Affiliation in Primates, a Model

This paper [link](https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1000630) aims to explain the complex and interesting social behaviours exhibited by Primates, such as reciprocal grooming, reconciling after a fight, and grooming those of similar status. The authors note existing theories, but suggest they are difficult to combine, suffer from a lack of evidence, and so on. They begin by listing the key observations and current hypotheses, which I will repeat briefly here.

## Background

Primates engage in both autogrooming (grooming yourself) and allogrooming (grooming someone else). There are many different explanations offered for the latter. Some assume it serves a hygenic purpose by removing ticks, fleas, etc. To back this up it has been observed that allogrooming tends to target regions that are inaccessible to the recipient (e.g., the head, back). On the the other hand it is pointed out that the amount of time spend doing this and the complexity of the social interactions are excessive. Interestingly those with a higher rank do less grooming than those with a lower rank, and most grooming occurs between pairs of similar rank. The consensus is now that grooming also serves some social role. As additional evidence for this it is noted that the time an individual spends grooming increases with the size of the group, but not with the body mass of the species. Grooming is known to causes relaxation in the donors and recipients, perhaps indicating a role in preventing conflict, and is also known to occur often between opponents after a fight. 

One prominent theory of grooming patterns was put forward by Robert Seyfarth ([original paper](https://bpb-us-w2.wpmucdn.com/web.sas.upenn.edu/dist/0/236/files/2017/04/seyfarth-jtheorbiol-zbkmjd.pdf)) to try and explain both the excess grooming of those with similar status (compared to random matchings), as well as the increased receivership with rank. His proposal was to have two factors, a target amount of time spent receiving grooming, $A_r$, as well as a target amount of grooming to give, $A_g$, which were common across the population. Grooming parterns are competitive, and those with the higher ranks get the first pick. Therefore, a greedy solution can be used and the model basically goes through each individual from highest rank to lowest. When someone has been groomed $A_r$ they cannot receive any more, and the groomer has to move on. The top individual is the first to choose who to groom and faces no competition. They will first choose the number 2, and so on. Here is a simple simulation:

```python
def simulate_Seyfarth(A_r, A_g, N):
    R = np.zeros(N)
    G = np.zeros(N)
    edges = []
    for i in range(N):
        for j in range(N):
            if not i == j:
                amount = min(A_r - R[j], A_g - G[i])
                if amount > 0:
                    edges.append((i, j))
                    R[j] += amount
                    G[i] += amount
    return R, G, edges

def plot_network(R, G, edges, N, A_r, A_g):
    G_nx = nx.DiGraph()
    G_nx.add_nodes_from(range(N))
    G_nx.add_edges_from(edges)
    
    angle_offset = - np.pi / 2
    positions = {
        i: (np.cos(2 * np.pi * i / N - angle_offset),
            np.sin(2 * np.pi * i / N - angle_offset))
        for i in range(N)
    }
    node_sizes = [1000 - 80 * i for i in range(N)]
    
    plt.figure(figsize=(3, 3))
    nx.draw(G_nx, pos=positions, with_labels=False, node_color='orange', 
            edge_color='black', node_size=node_sizes, arrows=True)
    
    for i in range(N):
        x, y = positions[i]
        label = f"R:{R[i]:.1f}\nG:{G[i]:.1f}"
        plt.text(x, y, label, ha='center', va='center', fontsize=9)
    
    plt.axis('off')
    plt.title(f"Grooming, A_r: {A_r}, A_g: {A_g}")
    plt.show()
```

```python
A_r = 16
A_g = 4
N = 10
R, G, edges = simulate_Seyfarth(A_r, A_g, N)
plot_network(R, G, edges, N, A_r, A_g)
```

The top individual is the highest rank and the ranks decrease anticlockwise. Seyfarth suggested that if the ratio of $A_r$ to $A_g$ was close to 1 then they would naturally end up grooming similar ranks:

```python
A_r = 11
A_g = 9
N = 10
R, G, edges = simulate_Seyfarth(A_r, A_g, N)
plot_network(R, G, edges, N, A_r, A_g)
```

Nice! So they are grooming those with close rank and those with lower rank don't receive much grooming. Seyfarth went on to describe other extensions of this idea, assuming that the individuals had different amount of spare time was one option, and caring more about family members was another. However, the model was eventually disputed by observations of Baboons ([link](https://rahill.webspace.durham.ac.uk/wp-content/uploads/sites/207/2021/04/Henzi-et-al-2003.pdf)), which found that there wasn't actually competition for grooming partners (no priority of access), and there was no evidence of a particular 'target' amount of grooming.

Another complicating factor is the evidence that the amount of grooming also corresponded to physical resources. Various biological market approaches were put forward to account for this, looking at it from the angle of partner selection.

The paper lists a few flaws with existing models (like above):
1. Lack of evidence, or contradictory evidence
2. No unity between different theories
3. The idea of markets / exchange is "dangerously antropomorphic".
4. Current models have primates making decisions based on a huge number of considerations, which is computationally difficult.

The authors of this paper then propose a new model, which they claim is more unified and computationally simple.
